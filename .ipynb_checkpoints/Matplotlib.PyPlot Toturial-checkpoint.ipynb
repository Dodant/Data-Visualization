## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graphs

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph
plt.figure(figsize=(5,3), dpi=300)

## Line 1

# shorthand notation
# plt.plot(x, y, 'r^--', label='2x')

plt.plot(x, y, label='2x', 
         color='blue', linewidth=2, linestyle='--', 
         marker='.', markersize=10, markeredgecolor='blue')

## Line 2
# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

# Add a Title
plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize' : 15})

# X and Y Labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y Tickmarks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12,14,16])

plt.legend()

plt.savefig('TestGraph.png', dpi=300)

plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

plt.show()
```
